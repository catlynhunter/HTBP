## Part 2: Comparison Operators - Comparing Values

Comparison operators let us **compare** two values. They always return a Boolean (`True` or `False`).

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `7 > 3` | `True` |
| `<` | Less than | `3 < 7` | `True` |
| `>=` | Greater than or equal to | `5 >= 5` | `True` |
| `<=` | Less than or equal to | `3 <= 5` | `True` |

### âš ï¸ IMPORTANT: `=` vs `==`

- **`=`** is for **assignment** (storing a value)
  - Example: `age = 20` â†’ Store 20 in the variable age

- **`==`** is for **comparison** (checking if equal)
  - Example: `age == 20` â†’ Check if age equals 20 (returns True or False)

````python
# Comparison examples
x = 10
y = 5

# Equal to
print(x == y)      # False bc (10 is not equal to 5)
print(x == 10)     # True bc (10 equals 10)

# Not equal to
print(x != y)      # True bc (10 is not equal to 5)

# Greater than
print(x > y)       # True bc (10 is greater than 5)

# Less than
print(x < y)       # False bc (10 is not less than 5)

# Greater than or equal to
print(x >= 10)     # True bc (10 is equal to 10)

# Less than or equal to
print(y <= 5)      # True bc (5 is equal to 5)
````

````python
# You can also compare strings!
name1 = "Alice"
name2 = "Bob"

print(name1 == "Alice")
print(name1 == "alice")   # False (case-sensitive!) - Alice is not the same as alice
print(name1 != name2)     # True bc Alice is not Bob, so this comes out as true

# Strings can be compared alphabetically
print("Apple" < "Banana")  # True (A comes before B)
````

### Boolean Values

Comparisons return Boolean values (`True` or `False`):

````python
# Store comparison results
age = 20
is_adult = age >= 18      # True
is_teenager = age < 18    # False

print("Is adult?", is_adult)
print("Is teenager?", is_teenager)
print("Type:", type(is_adult))  # <class 'bool'>
````

## Part 3: Logical Operators - Combining Conditions

Logical operators let us combine multiple conditions:

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `and` | Both must be True | `True and True` | `True` |
| `or` | At least one must be True | `True or False` | `True` |
| `not` | Reverses the boolean | `not True` | `False` |

### The `and` Operator

`and` returns `True` only if **BOTH** conditions are true.

````python
# Basic and examples
print(True and True)      # True
print(True and False)     # False
print(False and False)    # False

# Practical example
age = 20
has_id = True

# Can enter club if: age >= 18 AND has ID
can_enter_club = age >= 18 and has_id
print("Can enter club?", can_enter_club)

# Both conditions must be true
age = 16
has_id = True
can_enter_club = age >= 18 and has_id
print("Can enter club?", can_enter_club)  # False (age too young)
````

### The `or` Operator

`or` returns `True` if **AT LEAST ONE** condition is true.

````python
# Basic or examples
print(True or True)       # True
print(True or False)      # True
print(False or False)     # False

# Practical example
is_weekend = True
is_holiday = False

# Can sleep in if: weekend OR holiday
can_sleep_in = is_weekend or is_holiday
print("Can sleep in?", can_sleep_in)  # True

# Only need ONE to be true
is_weekend = False
is_holiday = True
can_sleep_in = is_weekend or is_holiday
print("Can sleep in?", can_sleep_in)  # Still True!
````

### The `not` Operator

`not` **reverses** a boolean value.

````python
# Basic not examples
print(not True)      # False
print(not False)     # True

# Practical example
is_raining = False
need_umbrella = not is_raining
print("Need umbrella?", need_umbrella)  # True (because it's NOT raining)

# Another example
has_homework = True
is_free = not has_homework
print("Are you free?", is_free)  # False (you have homework!)
````

### Combining Multiple Conditions

````python
# Complex example: Passing a class
score = 85
attendance = 90

# Pass if: score >= 60 AND attendance >= 75
passed = score >= 60 and attendance >= 75
print("Did you pass?", passed)  # True

# Even more complex
has_extra_credit = False
passed_or_extra = (score >= 60 and attendance >= 75) or has_extra_credit
print("Passed (considering extra credit)?", passed_or_extra)
````

## Part 4: If Statements - Making Decisions

Now we can use conditions to make our programs **make decisions**!

### Basic `if` Statement

```python
if condition:
    # Code to run if condition is True
    # This code must be indented!
```

````python
# Simple if statement
temperature = 75

if temperature > 70:
    print("It's warm outside!")
    print("Wear shorts!")

print("This line always runs")  # Not indented, so it will run regardless
````

````python
# More examples
age = 20

if age >= 18:
    print("You can vote!")

score = 95

if score >= 90:
    print("Excellent work!")
    print("You got an A!")
````

### `if-else` Statement

Do one thing if True, another thing if False:

```python
if condition:
    # Run if True
else:
    # Run if False
```

````python
# If-else example
age = 16

if age >= 18:
    print("You can vote!")
else:
    print("You cannot vote yet.")
    years_until = 18 - age
    print(f"Wait {years_until} more years.")
````

````python
# Another example
password = "python123"
user_input = input("Enter password: ")

if user_input == password:
    print("âœ“ Access granted!")
else:
    print("âœ— Access denied!")
````

### `if-elif-else` Statement

Check multiple conditions:

```python
if condition1:
    # Run if condition1 is True
elif condition2:
    # Run if condition1 is False AND condition2 is True
elif condition3:
    # Run if condition1 and condition2 are False AND condition3 is True
else:
    # Run if all conditions are False
```

````python
# Grade calculator
score = 85

if score >= 90:
    print("Grade: A")
    print("Excellent!")
elif score >= 80:
    print("Grade: B")
    print("Great job!")
elif score >= 70:
    print("Grade: C")
    print("Good work!")
elif score >= 60:
    print("Grade: D")
    print("You passed.")
else:
    print("Grade: F")
    print("Keep trying!")
````

````python
# Time of day greeting
hour = int(input("What hour is it? (0-23): "))

if hour < 12:
    print("Good morning!")
elif hour < 17:
    print("Good afternoon!")
elif hour < 21:
    print("Good evening!")
else:
    print("Good night!")
````

### Nested If Statements

You can put if statements inside other if statements:

````python
# Nested if example: Driving eligibility
age = 25
has_license = True

if age >= 16:
    if has_license:
        print("âœ“ You can drive!")
    else:
        print("âœ— You need a license to drive.")
else:
    print("âœ— You're too young to drive.")
    print(f"  Wait {16 - age} more years.")
````

### ðŸ”´ CRITICAL: Indentation in Python

**Python uses indentation (spaces) to define code blocks!**

Rules:
- Use **4 spaces** or **1 tab** for each level of indentation
- Be consistent (don't mix tabs and spaces)
- Everything at the same indentation level is part of the same block
- Wrong indentation = Error!

````python
# CORRECT indentation
x = 10

if x > 5:
    print("x is big")      # Indented - part of if block
    print("Very big!")     # Also indented - part of if block

print("Done")              # Not indented - always runs

# INCORRECT - will cause IndentationError
# if x > 5:
# print("This won't work")  # Not indented!
````
